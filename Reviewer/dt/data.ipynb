```python
import json

try:
    with open("ETL_OrbDataset.json", "r", encoding="utf-8") as f:
        data = json.load(f)
    print("JSON válido, keys:", list(data.keys()))
except json.JSONDecodeError as e:
    print(f"Error en línea {e.lineno}, columna {e.colno}")
    print(f"Mensaje: {e.msg}")
except UnicodeDecodeError as e:
    print(f"Problema de encoding: {e}")
```
