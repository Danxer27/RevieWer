```python
import chromadb
from pathlib import Path
DB_PATH = Path.cwd() / "sira_chroma_db"
cliente = chromadb.PersistentClient(path=str(DB_PATH))
collection = cliente.get_collection("checklist_guides")
```

```python
query_text = """"
2 Materials and methods
2.1 Data source and retrieval strategy
The bibliometric dataset was constructed through a structured search conducted in the
Scopus database. The search targeted publications related to the broad ﬁeld of Quan-
tum Computing, while systematically incorporating associated subdomains in quantum
information, communication, cryptography, sensing, metrology, hardware, simulation,
and software development to comprehensively capture the multidisciplinary landscape
of quantum technologies. The time span covered publications from 1980 through May
11, 2025. No language restrictions were applied. Eligible document types included journal
articles, reviews, book chapters, books, and conference proceedings.
To ensure high speciﬁcity and relevance, the search query was applied to the TITLE-
ABS-KEY ﬁeld using a complex Boolean strategy designed to retrieve publications di-
rectly addressing quantum computing and its technological ecosystem. A set of exclusion
ﬁlters was implemented to eliminate documents primarily focused on unrelated mate-
rials science or condensed matter physics topics where “quantum” is often used in non-
computational contexts. Speciﬁcally, documents referencing neutron scattering, polycrys-
talline materials, thin ﬁlms, boundary conditions, or chemical reaction theories were ex-
cluded.
The search expression was organized into seven thematic modules representing: (1)
quantum computing and quantum-mechanical computational models; (2) quantum in-
formation science, technology, and communication; (3) quantum cryptography and post-
quantum cryptographic approaches; (4) quantum sensing and quantum control; (5) quan-
tum metrology, tomography, clocks, gravimetry, and imaging; (6) quantum hardware sys-
tems and device architectures, and (7) quantum simulation, algorithms, and software
frameworks. These thematic modules reﬂect the conceptual structure of the search design.
The full Scopus query, including all Boolean operators, truncation rules, proximity oper-
ators, and ﬁeld speciﬁcations exactly as implemented, is provided in the Supplementary
Materials (Table S1). The complete search protocol, including scripts, search strings, and
preprocessing pipelines, is publicly archived at: https://github.com/DanielCondeTorres/"""
```

```python
collection.query(query_texts=query_text)
```

```python
 

# collection = cliente.get_collection("scipost_reviews")


where = None
results = collection.query(
        query_texts=[query_text],
        n_results=3,
        where=where,
        include=["documents", "metadatas", "distances"]
    )

print(results)
```
